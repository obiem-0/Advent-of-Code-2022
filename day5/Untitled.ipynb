```python
stacks = dict()

print(file.split("\n\n")[0])

for i in range(1, anr+1):
    
    stacks[i] = [n for n in [r[i-1] for r in rowsnew] if n != '#']
print(stacks)
```

```python
import re

for x in instr:
    
    steps = int(re.findall("move "+"(.*)"+" from", x)[0])
    
    start = int(re.findall("from "+"(.*)"+" to", x)[0])
    end = int(x.split("to ")[1])
    
    #get the characters whose index are less than the step and then reverse them  
    
    na = [v for k, v in enumerate(stacks[start]) if k <= steps]
    nb = stacks[end]

    print(steps, start, end)
    print ("stack " + str(start) + " before change: " + str(stacks[start]))
    print ("stack " + str(end) + " before change: " + str(stacks[end]))

    na.reverse()

    stacks[start] = [v for k, v in enumerate(stacks[start]) if k > steps]
    stacks[end] = na + nb
  
    print ("stack " + str(start) + " after change: " + str(stacks[start]))
    print ("stack " + str(end) + " after change: " + str(stacks[end]))
```
